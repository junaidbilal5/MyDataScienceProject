```python
import pandas as pd
import numpy as np
import sqlite3 as sql
```

```python
conn = sql.connect("chinook.db")
```

```python
def read_sql(q):
    return pd.read_sql_query(q, conn)
```

```python
q = """select * from sqlite_master"""
read_sql(q)
```

```python
q = """select * from invoice"""

read_sql(q)
```

```python
q = """select * from invoice_line where invoice_id = 1
        """

read_sql(q)
```

```python
q = """select i.invoice_id, i.billing_address,il.track_id, il.unit_price from invoice i
        inner join invoice_line il
        on 
        i.invoice_id = il.invoice_id
        where i.invoice_id = 1"""

read_sql(q)
```

```python
q = """pragma table_info(invoice)"""

read_sql(q)
```

```python
q = """select alb.title Album_Title, ar.name Artist_Name,
        count(il.track_id) No_Purchased 
        from artist ar
        left join album alb
        on
        alb.artist_id = ar.artist_id
        inner join track t
        on
        t.album_id = alb.album_id
        inner join invoice_line il
        on
        t.track_id = il.track_id
        group by 1
        order by 3 desc"""

read_sql(q)
```

```python
q = """Select e1.first_name ||' '|| e1.last_name Employee_Name, e1.title Employee_Title,
        e2.first_name ||' '|| e2.last_name Manager_Name, e2.title Manager_Title
        from employee e1
        left join employee e2
        on
        e1.reports_to = e2.employee_id
        """

read_sql(q)
```

```python
q = """select * from employee"""
read_sql(q)
```

```python
q = """select c.first_name || ' ' || c.last_name Customer_Name, c.country Countries,
        sum(inv.total) Total_spend
        from customer c
        inner join invoice inv
        on
        c.customer_id = inv.customer_id
        group by 1"""

read_sql(q)
```

```python
q = """with total_purchases as 
        (select c.first_name || ' ' || c.last_name Customer_Name, c.country Countries,
        sum(inv.total) Total_spend
        from customer c
        inner join invoice inv
        on
        c.customer_id = inv.customer_id
        group by 1)
        
        select Countries,Customer_Name,max(Total_spend) Top_Spend from total_purchases
        group by 1
        order by 3 desc
        """

read_sql(q)
```

```python
q = """select c.first_name || ' ' || c.last_name Customer_Name, c.country Countries,
        count(inv.invoice_id) no_of_purchases,
        sum(inv.total) Total_spend,
        
        case
        when sum(inv.total) < 40 then "Small Spender"
        when sum(inv.total) > 100 then "Big Spender"
        else "Regular"
        end as Spender
        
        from customer c
        inner join invoice inv
        on
        c.customer_id = inv.customer_id
        group by 1"""

read_sql(q)
```

```python
q = """with USA_Sale as
        (select g.name Genre_Name, count(t.track_id) Total_Sales from genre g
        inner join track t
        on 
        g.genre_id = t.genre_id
        inner join invoice_line il
        on
        t.track_id = il.track_id
        inner join invoice i
        on
        il.invoice_id = i.invoice_id
        inner join customer c
        on
        c.customer_id = i.customer_id
        where c.country == "USA"
        group by 1
        order by 2 desc)
        --- select sum(Total_Sales) from USA_sale;
        select genre_name, total_sales, 
        (cast(total_sales as float)/(select sum(Total_Sales) from USA_sale))*100 Percentage_Sale
        from usa_sale
        limit 10;
        """

df = read_sql(q)
df
```

```python
q = """with sales_table as
        (select e.first_name || ' ' || e.last_name Employee_Name, e.hire_date Hire_Date,
        i.total Total
        from employee e
        inner join customer c
        on
        e.employee_id = c.support_rep_id
        inner join invoice i
        on
        c.customer_id = i.customer_id)
        
        select Employee_name,Hire_Date,sum(Total) Total_sales from sales_table
        group by 1
        order by 3 desc
        """

read_sql(q)
```
